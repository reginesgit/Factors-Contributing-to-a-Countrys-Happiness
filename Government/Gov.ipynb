```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
Gov_file= "Resources/Database_of_Political_Institutions_2017.csv"
```

```python
# Read Data files with the pandas library
Gov_file_df = pd.read_csv(Gov_file, encoding="UTF-8")
```

```python
Gov_file_df
```

```python
Gov_file_df.head()
```

```python
# Rename Columns to match the datasets for previous years
Gov_file_df.rename(columns = {"countryname" : "Country",
                              "system" : "Form of Government",}, inplace = True)
                              
Gov_file_df.head()
```

```python
drop = Gov_file_df.columns[4:] 
drop
```

```python
Gov_file_df = Gov_file_df.drop(columns=drop, axis=1)
drop = Gov_file_df.columns[1] 
Gov_file_df = Gov_file_df.drop(columns=drop, axis=1)
Gov_file_df
```

```python
gov_new_dates = Gov_file_df.copy([Gov_file_df["year"] > 2014])
```

```python
gov_new_dates["Government"] = ""

gov_new_dates.loc[gov_new_dates["Form of Government"] == 0.0, "Government"] = "Direct Presidential"
gov_new_dates.loc[gov_new_dates["Form of Government"] == 1.0, "Government"] = "Strong president elected by assembly"
gov_new_dates.loc[gov_new_dates["Form of Government"] == 2.0, "Government"] = "Parliamentary"
```

```python
new_gov = gov_new_dates[gov_new_dates["year"] >= 2014]
```
