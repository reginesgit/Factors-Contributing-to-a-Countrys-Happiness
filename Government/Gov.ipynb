```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Store filepath in a variable
Gov_file= "Resources/Database_of_Political_Institutions_2017.csv"
```

```python
# Read Data files with the pandas library
Gov_file_df = pd.read_csv(Gov_file, encoding="UTF-8")
Gov_file_df
```

```python
# Print out Column Titles
Gov_file_df.columns
```

```python
# Remove columns not being used for analysis
Gov_file_df.drop(columns = ['ifs', 'yrsoffc', 'finittrm',
       'yrcurnt', 'multpl', 'military', 'defmin', 'percent1', 'percentl',
       'prtyin', 'execme', 'execrlc', 'execnat', 'execrurl', 'execreg',
       'execrel', 'execage', 'allhouse', 'nonchief', 'totalseats', 'gov1me',
       'gov1seat', 'gov1vote', 'gov1rlc', 'gov1nat', 'gov1rurl', 'gov1reg',
       'gov1rel', 'gov1age', 'gov2me', 'gov2seat', 'gov2vote', 'gov2rlc',
       'gov2nat', 'gov2rurl', 'gov2reg', 'gov2rel', 'gov2age', 'gov3me',
       'gov3seat', 'gov3vote', 'gov3rlc', 'gov3nat', 'gov3rurl', 'gov3reg',
       'gov3rel', 'gov3age', 'govoth', 'govothst', 'govothvt', 'opp1me',
       'opp1seat', 'opp1vote', 'opp1rlc', 'opp1nat', 'opp1rurl', 'opp1reg',
       'opp1rel', 'opp1age', 'opp2me', 'opp2seat', 'opp2vote', 'opp3me',
       'opp3seat', 'opp3vote', 'oppoth', 'oppothst', 'oppothvt', 'ulprty',
       'numul', 'ulvote', 'oppmajh', 'oppmajs', 'dateleg', 'dateexec',
       'legelec', 'exelec', 'liec', 'eiec', 'mdmh', 'mdms', 'ssh', 'pluralty',
       'pr', 'housesys', 'sensys', 'thresh', 'dhondt', 'cl', 'select', 'fraud',
       'auton', 'muni', 'state', 'author', 'stconst', 'gwno'], inplace = True)
Gov_file_df
```

```python
# Only keep rows for year 2017
Gov_file_df = Gov_file_df[Gov_file_df['year']== 2017]
Gov_file_df
```

```python
# Rename Columns 
Gov_file_df.rename(columns = {"countryname" : "Country",
                              "system" : "Form of Government",}, inplace = True)
                              
Gov_file_df
```

```python
# Find Invalid forms of Governments 
invalid = Gov_file_df["Form of Government"] == -999

#Drop Invalid Goverments by index -  inplace 

print(invalid)

Gov_file_df.drop(Gov_file_df[invalid].index, inplace = True)
```

```python
# Drop all rows with 'NaN'
Gov_file_df = Gov_file_df.dropna()
Gov_file_df.reset_index(inplace = True, drop = True)
Gov_file_df
```

```python
# Presidential (0), chief executive is chosen by the people to serve a limited term in office with a distinct 
# separation of powers (the executive branch) as well as specific limitations on exactly what he/she can do while 
#in office. Systems with presidents who are elected directly or by an electoral college 
#(whose only function is to elect the president) and in cases where there is no prime minister
# Assembly-elected President (1) any means of executive selection not involving A Direct Or Indirect mandate from An electorate
#when that assembly or group cannot easily recall him 
# Parliamentary (2), Countries in which the legislature elects the chief executive. 
#in this system, the executive and legislative branches of government are combined and the political head of state 
#is chosen from within the legislature
```

```python


# replace form of goverment code to string
Gov_file_df["Government"] = ""

Gov_file_df.loc[Gov_file_df["Form of Government"] == 0.0, "Government"] = "Direct Presidential"
Gov_file_df.loc[Gov_file_df["Form of Government"] == 1.0, "Government"] = "Assembly Elected President"
Gov_file_df.loc[Gov_file_df["Form of Government"] == 2.0, "Government"] = "Parliamentary"
Gov_file_df
```

```python
#save cleaned government data as csv
Gov_file_df
Gov_file_df.to_csv("Resources/cleaned_gov_file.csv")
```

```python
file = "../Happiness/Resources/avg_happiness.csv"
merge_ave_happiness = pd.read_csv(file)
merge_ave_happiness



```

```python
#merge happiness data with Government data by Country

merge_happiness_Gov = pd.merge(merge_ave_happiness, Gov_file_df, on="Country", how="left")
merge_happiness_Gov = merge_happiness_Gov.dropna()
merge_happiness_Gov.reset_index(inplace = True, drop = True)
merge_happiness_Gov
```

```python
#save new file as csv
merge_happiness_Gov
merge_happiness_Gov.to_csv("Resources/merge_happiness_gov.csv")
```

```python
# DataFrom for all the SAME Governments

par = merge_happiness_Gov[merge_happiness_Gov["Government"] == "Parliamentary"]
direct = merge_happiness_Gov[merge_happiness_Gov["Government"] == "Direct Presidential"]
strong = merge_happiness_Gov[merge_happiness_Gov["Government"] == "Assembly Elected President"]

par.mean()
direct.mean()
strong.mean()
```

```python
# Getting Happiness Mean by form of government  
par_avg_happy = par["Avg Happiness"].mean()
direct_avg_happy = direct["Avg Happiness"].mean()
strong_avg_happy = strong["Avg Happiness"].mean()


avg_happiness = {"Governments" : ["Parliamentary","Direct Presidential","Assembly Elected President"],
                 "Average Happiness" : [par_avg_happy, direct_avg_happy, strong_avg_happy] }


print(par_avg_happy)
print(direct_avg_happy)
print(strong_avg_happy)
```

```python
# plot the average happiness rws
avg_by_government = pd.DataFrame.from_records(avg_happiness)
avg_by_government.plot.bar(y="Average Happiness", x="Governments")

plt.title("Hapiness Mean by form of Government", fontweight='bold', color = 'black', fontsize='15', horizontalalignment='center')
plt.xlabel("Form of Government", fontweight='bold', color = 'black', fontsize='12', horizontalalignment='center')
plt.ylabel("Happiness Mean", fontweight='bold', color = 'black', fontsize='12', horizontalalignment='center')
plt.legend(loc="upper right")

```
