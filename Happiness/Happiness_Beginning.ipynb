```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
file_2015 = "Resources/2015.csv"
file_2016 = "Resources/2016.csv"
file_2017 = "Resources/2017.csv"
file_2018 = "Resources/2018.csv"
file_2019 = "Resources/2019.csv"
```

```python
# Read Data files with the pandas library
happiness_2015_df = pd.read_csv(file_2015, encoding="UTF-8")
happiness_2016_df = pd.read_csv(file_2016, encoding="UTF-8")
happiness_2017_df = pd.read_csv(file_2017, encoding="UTF-8")
happiness_2018_df = pd.read_csv(file_2018, encoding="UTF-8")
happiness_2019_df = pd.read_csv(file_2019, encoding="UTF-8")
```

```python
# Drop columns missing from other years
happiness_2015_df.drop(columns = ["Region", "Standard Error", "Dystopia Residual"], inplace = True)
```

```python
# Rename Columns to match the datasets for previous years and add suffix for year
happiness_2015_df.rename(columns = {"Happiness Rank" : "Happiness Rank_2015",
                              "Happiness Score" : "Happiness Score_2015",
                              "Economy (GDP per Capita)" : "Economy (GDP per Capita)_2015",
                              "Family" : "Family_2015",
                              "Health (Life Expectancy)" : "Health (Life Expectancy)_2015",
                              "Freedom" : "Freedom_2015",
                              "Trust (Government Corruption)" : "Trust (Government Corruption)_2015",
                              "Generosity" : "Generosity_2015"}, inplace = True)
happiness_2015_df.head()
```

```python
# Drop columns missing from other years
happiness_2016_df.drop(columns = ["Region", 'Lower Confidence Interval', 'Upper Confidence Interval', "Dystopia Residual"], inplace = True)
```

```python
# Rename Columns to match the datasets for previous years and add suffix for year
happiness_2016_df.rename(columns = {"Happiness Rank" : "Happiness Rank_2016",
                              "Happiness Score" : "Happiness Score_2016",
                              "Economy (GDP per Capita)" : "Economy (GDP per Capita)_2016",
                              "Family" : "Family_2016",
                              "Health (Life Expectancy)" : "Health (Life Expectancy)_2016",
                              "Freedom" : "Freedom_2016",
                              "Trust (Government Corruption)" : "Trust (Government Corruption)_2016",
                              "Generosity" : "Generosity_2016"}, inplace = True)
happiness_2016_df.head()
```

```python
# Drop columns missing from other years
happiness_2017_df.drop(columns = ["Whisker.low", "Whisker.high", "Dystopia.Residual"], inplace = True)
```

```python
# Rename Columns to match the datasets for previous years and add suffix for year
happiness_2017_df.rename(columns = {"Happiness.Rank" : "Happiness Rank_2017",
                              "Happiness.Score" : "Happiness Score_2017",
                              "Economy..GDP.per.Capita." : "Economy (GDP per Capita)_2017",
                              "Family" : "Family_2017",
                              "Health..Life.Expectancy." : "Health (Life Expectancy)_2017",
                              "Freedom" : "Freedom_2017",
                              "Trust..Government.Corruption." : "Trust (Government Corruption)_2017",
                              "Generosity" : "Generosity_2017"}, inplace = True)
happiness_2017_df.head()
```

```python
# Rename Columns to match the datasets for previous years and add suffix for year
happiness_2018_df.rename(columns = {"Overall rank" : "Happiness Rank_2018",
                                    "Country or region": "Country",
                              "Score" : "Happiness Score_2018",
                              "GDP per capita" : "Economy (GDP per Capita)_2018",
                              "Social support" : "Family_2018",
                              "Healthy life expectancy" : "Health (Life Expectancy)_2018",
                              "Freedom to make life choices" : "Freedom_2018",
                              "Perceptions of corruption" : "Trust (Government Corruption)_2018",
                              "Generosity" : "Generosity_2018"}, inplace = True)
happiness_2018_df.head()
```

```python
# Rename Columns to match the datasets for previous years
happiness_2019_df.rename(columns = {"Overall rank" : "Happiness Rank_2019",
                              "Score" : "Happiness Score_2019",
                              "GDP per capita" : "Economy (GDP per Capita)_2019",
                              "Social support" : "Family_2019",
                              "Healthy life expectancy" : "Health (Life Expectancy)_2019", 
                              "Freedom to make life choices" : "Freedom_2019",
                              "Perceptions of corruption" : "Trust (Government Corruption)_2019",
                              "Country or region" : "Country",
                                   "Generosity" : "Generosity_2019"}, inplace = True)
happiness_2019_df.head()
```

```python
# Merge DataFrame from 2015 and 2016
happiness_15_16_df = happiness_2015_df.merge(happiness_2016_df, how = "left", on = "Country")
happiness_15_16_df.head()
```

```python
# Merge DataFrame to 2017
happiness_15_17_df = happiness_15_16_df.merge(happiness_2017_df, how = "left", on = "Country")
happiness_15_17_df.head()
```

```python
# Merge DataFrame to 2018
happiness_15_18_df = happiness_15_17_df.merge(happiness_2018_df, how = "left", on = "Country")
happiness_15_18_df.head()
```

```python
# Merge DataFrame to 2019
happiness_15_19_df = happiness_15_18_df.merge(happiness_2019_df, how = "left", on = "Country")
happiness_15_19_df.to_csv("Resources/merged_happiness_data.csv")
happiness_15_19_df.head()
```

```python
# Create DataFrame with Happiness score per date
avg_happiness_df = happiness_15_19_df[["Country", "Happiness Score_2015", "Happiness Score_2016", "Happiness Score_2017",
                                      "Happiness Score_2018", "Happiness Score_2019"]]
avg_happiness_df
```

```python
# Drop rows with 'NaN'
avg_happiness_df = avg_happiness_df.dropna()
```

```python
# Calculate Avg. Happiness and create a new column
avg_happiness_df['Avg Happiness'] = (avg_happiness_df["Happiness Score_2015"] + 
                           avg_happiness_df["Happiness Score_2016"]+ 
                           avg_happiness_df["Happiness Score_2017"]+ 
                           avg_happiness_df["Happiness Score_2018"]+ 
                           avg_happiness_df["Happiness Score_2019"])/5

avg_happiness_df.to_csv("Resources/avg_happiness.csv")

avg_happiness_df
```
